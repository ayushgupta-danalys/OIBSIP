```python
#filling missing rows
df['name'].fillna('unknown',inplace=True)
df['host_name'].fillna('no host',inplace=True)
df['reviews_per_month'].fillna(0,inplace=True)
df['last_review'].fillna('no review',inplace=True)

df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#seen no duplicates still confirm for data validation

print("before removing duplicates :",df.shape)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
print("after removing duplicates :",df.shape)
```

```python
#standarize the columns
df['neighbourhood_group']=df['neighbourhood_group'].str.lower().str.strip()
df['neighbourhood']=df['neighbourhood'].str.lower().str.strip()
df['room_type']=df['room_type'].str.lower().str.strip()

#types of room and neighbourhood group
print("unique room types:", df['room_type'].unique())
print("unique neighbourhood groups: ",df['neighbourhood_group'].unique()) 
```

```python
#fixing data types
df['price']=df['price'].astype(float)
df['minimum_nights']=df['minimum_nights'].astype(int)
```

```python
#checking outliers
df=df[df['price']>0]
df.reset_index(drop=True, inplace=True)
df.shape[0]
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['price'])
plt.title('price boxplot - before removing outliers')
plt.show()
```

```python
# remove outliers using IQR method
Q1=df['price'].quantile(0.25)
Q3=df['price'].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
print("lower bound: ", lower)
print("upper bound: ",upper)

df_cleaned=df[(df['price']>=lower)&(df['price']<=upper)]
df_cleaned=df_cleaned.reset_index(drop=True)

print("shape before removing: ",df.shape)
print("shape before removing: ",df_cleaned.shape)
```

```python
#boxploting
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_cleaned['price'])
plt.title('price boxplot - after removing outliers')
plt.show()
```

```python
#price distribution
plt.figure(figsize=(10, 5))
plt.hist(df_cleaned['price'],bins=50,color='steelblue',edgecolor='black')
plt.title('price distribution')
plt.xlabel('price')
plt.ylabel('count')
plt.show()
```

```python
#neighbourhood group count
plt.figure(figsize=(8, 5))
df_cleaned['neighbourhood_group'].value_counts().plot(kind='bar',color='coral',edgecolor='black')
plt.title('listing by neighbourhood group')
plt.xlabel('neighbourhood group')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.show()
```

```python
#count by room type
plt.figure(figsize=(8, 5))
df_cleaned['room_type'].value_counts().plot(kind='bar',color='green',edgecolor='black')
plt.title('room type count')
plt.xlabel('room type')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.show()
```

```python
# summary
print("original shape: (48895, 16)")
print("cleaned_shape:",df_cleaned.shape)
print("missing values:",df_cleaned.isnull().sum())
print("duplicate:",df_cleaned.duplicated().sum())
print("price range:", df_cleaned['price'].min(), "to" , df_cleaned['price'].max())
```
