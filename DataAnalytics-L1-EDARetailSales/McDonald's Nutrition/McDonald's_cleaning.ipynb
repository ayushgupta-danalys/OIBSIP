```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
#strip whitespace from text columns
df['Category']=df['Category'].str.strip()
df['Item']=df['Item'].str.strip()
df.shape
```

```python
#serving size column has mixed values like 'fl oz' and 'oz'
#just checking what kind of values are in there
df['Serving Size'].head(15)
```

```python

print("avg calories per item: ",round(df['Calories'].mean(),2))
print("max calories item")
print(df[df['Calories']==df['Calories'].max()][['Item', 'Calories']])
print("min calories item")
print(df[df['Calories']==df['Calories'].min()][['Item', 'Calories']])
```

```python
print("avg nutrition per by category: ")
print(df.groupby('Category')['Calories'].mean().sort_values(ascending=False).round(2))
```

```python
print("avg protein by category :")
print(df.groupby('Category')['Protein'].mean().sort_values(ascending=False).round(2))
```

```python
#number of items per category
item_count= df['Category'].value_counts()
print("items per category:")
print(item_count)
```

```python
plt.figure(figsize=(10,5))
item_count.plot(kind='bar', color='steelblue',edgecolor='black')
plt.title('numbers of item per category')
plt.xlabel('category')
plt.ylabel('count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#avg calories by category
avg_cal= df.groupby('Category')['Calories'].mean().sort_values(ascending=False)


plt.figure(figsize=(10,5))
avg_cal.plot(kind='bar', color='orange',edgecolor='black')
plt.title('average calories by category')
plt.xlabel('category')
plt.ylabel('calories')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#top 10 highest calorite items
top_cal= df[['Item','Calories']].sort_values('Calories', ascending=False).head(10)
print(top_cal)
```

```python
plt.figure(figsize=(10,5))
plt.barh(top_cal['Item'],top_cal['Calories'],color='red',edgecolor='black')
plt.title('top 10 highest calories items')
plt.xlabel('Calories')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
#top 10 highest calorite items
top_protein= df[['Item','Protein']].sort_values('Protein', ascending=False).head(10)
print(top_protein)
```

```python
plt.figure(figsize=(10,5))
plt.barh(top_protein['Item'],top_protein['Protein'],color='green',edgecolor='black')
plt.title('top 10 highest protein items')
plt.xlabel('Protein(g)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
#correlation heatmap
plt.figure(figsize=(10, 7))
numeric_cols = df[['Calories', 'Total Fat', 'Sodium', 'Carbohydrates','Protein','Sugars']]
sns.heatmap(numeric_cols.corr(), annot=True, cmap='Oranges', fmt='.2f')
plt.title('Nutrition Correlation Heatmap')
plt.show()
```

```python
# calories vs protein scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Calories'], df['Protein'], alpha=0.5, color='steelblue')
plt.title('Calories vs Protein')
plt.xlabel('Calories')
plt.ylabel('Protein (g)')
plt.show()
# i noticed items with more calories tend to have more protein too
```

```python
# fat distribution
plt.figure(figsize=(8, 5))
plt.hist(df['Total Fat'], bins=30, color='salmon', edgecolor='black')
plt.title('Total Fat Distribution')
plt.xlabel('Total Fat (g)')
plt.ylabel('Count')
plt.show()
```

```python
# pie chart for category distribution
category_count = df['Category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(category_count.values, labels=category_count.index,autopct='%1.1f%%', startangle=140)
plt.title('Menu Item Distribution by Category')
plt.show()
```

```python
#summary
print("total menu items:", len(df))
print("total categories:", df['Category'].nunique())
print("highest calorie item:", df.loc[df['Calories'].idxmax(), 'Item'])
print("lowest calorie item:", df.loc[df['Calories'].idxmin(), 'Item'])
print("highest protein item:", df.loc[df['Protein'].idxmax(), 'Item'])
```

```python
# my observations
# 1. smoothies and shakes category has highest average calories
# 2. chicken and fish items have the highest protein content
# 3. calories and total fat are strongly correlated
# 4. breakfast items have high sodium content
# 5. salads have the lowest average calories which makes sense
print("done")
```
