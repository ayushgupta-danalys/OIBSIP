```python
## Feature Selection

#In this step, we will separate the input features and the target variable.

#The input features are the chemical properties of the wine, and the target variable is the quality of the wine.
```

```python
# Selecting input features and target variable

# 'Id' is only a unique identifier, so it is not useful for prediction.
# We will remove both 'Id' and 'quality' from the input features.

X = df.drop(["Id", "quality"], axis=1)

y = df["quality"]

print("Features Shape :", X.shape)
print("Target Shape :", y.shape)
```

```python
#The variable **X** contains all the input features, while **y** contains the wine quality that we want to predict.
```

```python
## Splitting the Dataset

#The dataset is divided into training and testing data.

#- Training Data = 80%
#- Testing Data = 20%
```

```python
# Import train test split

from sklearn.model_selection import train_test_split

# Splitting dataset

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Data :", X_train.shape)
print("Testing Data :", X_test.shape)
```

```python
## Feature Scaling

#Some machine learning algorithms perform better when the data is scaled.

#So we will standardize the feature values.
```

```python
# Import StandardScaler

from sklearn.preprocessing import StandardScaler

# Creating scaler object

scaler = StandardScaler()

# Scaling training and testing data

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
#The training data is used to calculate the scaling values.

#The same scaling is then applied to the testing data.
```

```python
# Random Forest Classifier

#Now we will train the Random Forest model.
```

```python
# Import Random Forest

from sklearn.ensemble import RandomForestClassifier

# Creating model

rf_model = RandomForestClassifier(random_state=42)
```

```python
# Training the model

rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully.")
```

```python
# Predicting wine quality

rf_pred = rf_model.predict(X_test)

rf_pred[:10]
```

```python
# Creating comparison table

rf_result = pd.DataFrame({
    "Actual Quality": y_test.values,
    "Predicted Quality": rf_pred
})

rf_result.head(10)
```

```python
# Import accuracy score

from sklearn.metrics import accuracy_score

# Calculate accuracy

rf_accuracy = accuracy_score(y_test, rf_pred)

print("Random Forest Accuracy :", rf_accuracy)
```

```python
#Higher accuracy means the model predicts wine quality more correctly.
```

```python
# Import confusion matrix

from sklearn.metrics import confusion_matrix

# Calculate confusion matrix

cm = confusion_matrix(y_test, rf_pred)

cm
```

```python
# Visualizing confusion matrix

plt.figure(figsize=(7,5))

sns.heatmap(cm,
            annot=True,
            fmt="d",
            cmap="Blues")

plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

```python
#The confusion matrix shows how many predictions were correct and where the model made mistakes
```

```python
# Import classification report

from sklearn.metrics import classification_report

# Display report

print(classification_report(y_test, rf_pred))
```

```python
#The classification report provides Precision, Recall, F1-Score, and Support for each wine quality class.

#These metrics help evaluate the model in more detail than accuracy alone.
```

```python
# Stochastic Gradient Descent (SGD) Classifier

#Now we will train the SGD Classifier and evaluate its performance.
```

```python
# Import SGD Classifier

from sklearn.linear_model import SGDClassifier

# Creating the model

sgd_model = SGDClassifier(random_state=42)
```

```python
# Training the SGD model

sgd_model.fit(X_train_scaled, y_train)

print("SGD model trained successfully.")
```

```python
# Predicting wine quality

sgd_pred = sgd_model.predict(X_test_scaled)

sgd_pred[:10]
```

```python
# Creating comparison table

sgd_result = pd.DataFrame({
    "Actual Quality": y_test.values,
    "Predicted Quality": sgd_pred
})

sgd_result.head(10)
```

```python
# Calculating accuracy

sgd_accuracy = accuracy_score(y_test, sgd_pred)

print("SGD Accuracy :", sgd_accuracy)
```

```python
#The accuracy score tells us how many predictions were correct.

#A higher accuracy means the model performs better.
```

```python
# Creating confusion matrix

sgd_cm = confusion_matrix(y_test, sgd_pred)

sgd_cm
```

```python
# Plot confusion matrix

plt.figure(figsize=(7,5))

sns.heatmap(sgd_cm,
            annot=True,
            fmt="d",
            cmap="Greens")

plt.title("SGD Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

```python
#The confusion matrix helps us understand where the SGD model made correct and incorrect predictions.
```

```python
# Display classification report

print(classification_report(y_test, sgd_pred))
```

```python
#The classification report shows Precision, Recall and F1-Score for every quality class.

#These values help us evaluate the overall performance of the SGD model.
```

```python
# Support Vector Classifier (SVC)

#Now we will train the Support Vector Classifier and compare its performance with the previous models.

#Since SVC works better with scaled data, we will use the scaled dataset.
```

```python
# Import Support Vector Classifier

from sklearn.svm import SVC

# Creating the model

svc_model = SVC(random_state=42)
```

```python
# Training the model

svc_model.fit(X_train_scaled, y_train)

print("SVC model trained successfully.")
```

```python
# Predicting wine quality

svc_pred = svc_model.predict(X_test_scaled)

svc_pred[:10]
```

```python
# Creating comparison table

svc_result = pd.DataFrame({
    "Actual Quality": y_test.values,
    "Predicted Quality": svc_pred
})

svc_result.head(10)
```

```python
# Calculating accuracy

svc_accuracy = accuracy_score(y_test, svc_pred)

print("SVC Accuracy :", svc_accuracy)
```

```python
#The accuracy score tells how well the SVC model predicts wine quality.
```

```python
# Creating confusion matrix

svc_cm = confusion_matrix(y_test, svc_pred)

svc_cm
```

```python
# Plotting confusion matrix

plt.figure(figsize=(7,5))

sns.heatmap(
    svc_cm,
    annot=True,
    fmt="d",
    cmap="Oranges"
)

plt.title("SVC Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```

```python
#The confusion matrix helps us identify correct and incorrect predictions made by the SVC model.
```

```python
# Display classification report

print(classification_report(y_test, svc_pred))
```

```python
#The classification report provides Precision, Recall and F1-Score for every class.

#These metrics give a better understanding of the model's performance.
```

```python
# Comparing model accuracy

comparison = pd.DataFrame({
    "Model": [
        "Random Forest",
        "SGD Classifier",
        "Support Vector Classifier"
    ],
    "Accuracy": [
        rf_accuracy,
        sgd_accuracy,
        svc_accuracy
    ]
})

comparison
```

```python
# Sorting models by accuracy

comparison.sort_values(
    by="Accuracy",
    ascending=False
)
```

```python
#The comparison table shows the accuracy of all three machine learning models.

#The model with the highest accuracy performed the best on this dataset.
```

```python
# Feature importance from Random Forest

importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf_model.feature_importances_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

importance
```

```python
# Plot feature importance

plt.figure(figsize=(10,6))

sns.barplot(
    data=importance,
    x="Importance",
    y="Feature"
)

plt.title("Feature Importance (Random Forest)")

plt.show()
```
